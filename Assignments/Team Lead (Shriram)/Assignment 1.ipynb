# Task 3 : Visualizations

## Univariate Analysis

#### Scatterplots of One Variable

```python
sns.scatterplot(x=data['id'],y=data['living area'])
```

### Distplot Of One Variable

```python
sns.histplot(data['Area of the house(excluding basement)'])
```

```python
sns.distplot(data['Area of the house(excluding basement)'])
```

### Visualizations of Grade Data

```python
sns.barplot(x=data['grade of the house'].value_counts().index,y=data['grade of the house'].value_counts())
```

## Bivariate Analysis

```python
sns.lineplot(x=data['grade of the house'],y=data['Area of the house(excluding basement)'])
```

```python
sns.lineplot(x=data['Area of the house(excluding basement)'],y=data['Price'])
```

```python
sns.scatterplot(x=data['number of bedrooms'],y=data['number of bathrooms'])
```

## Multivariate Analysis

```python
from matplotlib import pyplot
pyplot.figure(figsize=(30, 30))
sns.heatmap(data.corr(),annot=True)
```

```python
sns.pairplot(data[['number of bedrooms', 'number of bathrooms',
       'living area', 'number of floors', 'condition of the house', 'grade of the house',
       'Area of the house(excluding basement)','Number of schools nearby', 'Price']])
```
